```python
from abc import ABC, abstractmethod

class Runnable(ABC):
  @abstractmethod
  def invoke(input_data):
    pass
```

```python
import random

class NakliLLM:
  def __init__(self):
    print("LLM Created")


  def invoke(self, prompt): # dummmy response
    response = [
        "Delhi is the capital of India",
        "IPL is a cricket league",
        "AI stands for Artificial Intelligence"
    ]

    return {'response': random.choice(response)}


  def predict(self, prompt): # dummmy response
    response = [
        "Delhi is the capital of India",
        "IPL is a cricket league",
        "AI stands for Artificial Intelligence"
    ]

    return {'response': random.choice(response)}
```

```python
class NakliPromptTemplate(Runnable):
  def __init__(self, template, input_variables):
    self.template = template
    self.input_variables = input_variables


  def invoke(self, input):
    return self.template.format(**input)


  def format(self, input):
    return self.template.format(**input)
```

```python
class NakliStrOutputParser(Runnable):
  def invoke(self, input):
    return input['response']
```

```python
class RunnableConnector(Runnable):
  def __init__(self, runnable_list):
    self.runnable_list = runnable_list

  def invoke(self, input_data):
    for runnable in self.runnable_list:
      input_data = runnable.invoke(input_data)
    return input_data
```

```python
prompt1 = NakliPromptTemplate(
    template='Write a joke about a {topic}',
    input_variables=['topic']
)
```

```python
prompt2 = NakliPromptTemplate(
    template='Give explanation about a {response}',
    input_variables=['response']
)
```

```python
llm = NakliLLM()
```

```python
parser = NakliStrOutputParser()
```

```python
chain1 = RunnableConnector([prompt1,llm])
```

```python
chain2 = RunnableConnector([prompt2,llm,parser])
```

```python
final_chain = RunnableConnector([chain1,chain2])
final_chain.invoke({'topic':'cricket'})
```
