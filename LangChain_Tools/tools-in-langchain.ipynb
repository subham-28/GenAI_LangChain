## Custom Tools

```python
from langchain_core.tools import tool
```

```python
# step-1: create a func
def multiply(a,b):
  """Multiply two numbers"""
  return a*b
```

```python
# step-2: add type hints
def multiply(a:int,b:int) -> int:
  """Multiply two numbers"""
  return a*b
```

```python
# step-3: add a tool decorator

@tool
def multiply(a:int,b:int) -> int:
  """Multiply two numbers"""
  return a*b
```

```python
result=multiply.invoke({"a":2,"b":3})
print(result)
```

```python
print(multiply.name)
print(multiply.description)
print(multiply.args)
```

```python
# what goes inside llm
multiply.args_schema.model_json_schema()
```

## Method 2: Using Structured Tool

```python
from langchain_core.tools import StructuredTool
from pydantic import BaseModel, Field
```

```python
class MultiplyInput(BaseModel):
  a:int=Field(required=True,description="first number")
  b:int=Field(required=True,description="second number")
```

```python
def multiply_func(a:int,b:int) -> int:
  return a*b
```

```python
multiply_tool=StructuredTool.from_function(
    func=multiply_func,
    name="multiply",
    description="Multiply two numbers",
    args_schema=MultiplyInput
)
```

```python
result=multiply_tool.invoke({"a":2,"b":3})
print(result)
```

## Method-3: Using BaseTool Class

```python
from langchain_core.tools import BaseTool
from typing import Type
```

```python
class MultiplyInput(BaseModel):
  a:int=Field(required=True,description="first number")
  b:int=Field(required=True,description="second number")
```

```python
class MultiplyTool(BaseTool):
  name: str="multiply"
  description: str="Multiply two numbers"
  args_schema: Type[BaseModel]=MultiplyInput

  def _run(self,a:int,b:int) -> int:
    return a*b
```

```python
multiply_tool=MultiplyTool()
```

```python
result=multiply_tool.invoke({"a":2,"b":3})
print(result)
```

## Toolkit

```python
from langchain_core.tools import tool

# Custom tools
@tool
def add(a: int, b: int) -> int:
    """Add two numbers"""
    return a + b

@tool
def multiply(a: int, b: int) -> int:
    """Multiply two numbers"""
    return a * b
```

```python
class MathToolkit:
  def get_tools(self):
    return [add, multiply]
```

```python
toolkit = MathToolkit()
tools = toolkit.get_tools()

for tool in tools:
    print(tool.name, "=>", tool.description)
```
